```python
# Extracting data from the dataset

dataFile = open('materialsDataset.txt', 'r')
dataLines = dataFile.readlines()
cityList = []
peopleList = []
budgetList = []
papersList = []
efficiencyList = []
for i in dataLines[1:]:
    a = i.strip()
    b = a.split()
    cityList.append(b[0])
    peopleList.append(float(b[1]))
    budgetList.append(float(b[2]))
    papersList.append(float(b[3]))
    efficiencyList.append(str(b[4]))
    
#print (cityList)
#print (peopleList)
#print (budgetList)
```

```python
# Reshaping and joining data for K-Means analysis

people = np.asarray(peopleList).reshape(69, 1)
budget = np.asarray(budgetList).reshape(69, 1)
values = np.concatenate((people, budget), axis = 1)
```

```python
# Determining the optimal number of clusters using ELBOW method

distance = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 12)
    kmeans.fit(values)
    distance.append(kmeans.inertia_)
    
plt.plot(range(1, 11), distance)
plt.grid()
plt.title('Optimal number of clusters', fontsize = 18)
plt.xlabel('Number of clusters', fontsize = 15)
plt.ylabel('Distance from centroid', fontsize = 15)
plt.show()
```

```python
# Using K-Means algorithm with 2 clusters to separate our data into 2 groups

kmeans = KMeans(n_clusters = 2)
kmeans.fit(values)
y_kmeans = kmeans.predict(values)
```

```python
# Visualizing the clusters predicted by K-Means

plt.scatter(values[:, 0], values[:, 1], marker = 'x', c = y_kmeans, s = 20, cmap = 'Dark2')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'red', s = 100, alpha = 0.9)
plt.title('Predicted clusters', fontsize = 18)
plt.xlabel('Number of people working in materials science', fontsize = 15)
plt.ylabel('Annual budget (in Millions)', fontsize = 15)
plt.show()
```

```python
# Splitting our data into training and testing dataset for supervised machine learning

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(peopleList, budgetList, random_state = 1, test_size = 0.25)
```

```python
print (xtrain)
print (xtest)
print (ytrain)
print (ytest)
```

```python
# Transforming our data into numpy array

xtrain = np.asarray(xtrain)
xtest = np.asarray(xtest)
ytrain = np.asarray(ytrain)
#ytest = np.asarray(xtest)
ytest = np.asarray(ytest)
print (xtrain)
print (xtest)
print (ytrain)
print (ytest)
```

```python
# Applying gaussian naive bayes to our training dataset

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model = GaussianNB()
model.fit(xtrain.reshape(-1, 1), ytrain.reshape(-1, 1))
ymodel = model.predict(xtrain.reshape(-1, 1))
accuracy = round(model.score(xtrain.reshape(-1, 1), ytrain.reshape(-1, 1)) * 100, 2)
print (accuracy)
```

```python
# Using the trained model on testing dataset for prediction.

ymodel = model.predict(xtest.reshape(-1, 1))
print (ytest)
print (ymodel)
```

```python
# Using logistic regression on our training dataset

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xtrain.reshape(-1, 1), ytrain.reshape(-1, 1))
y_pred = logreg.predict(xtrain.reshape(-1, 1))
accuracy = round(logreg.score(xtrain.reshape(-1, 1), ytrain.reshape(-1, 1)) * 100, 2)
print (accuracy)
```

```python
y_pred = logreg.predict(xtest.reshape(-1, 1))
print (ytest)
print (y_pred)
```

```python
# Using perceptron model on our training dataset

from sklearn.linear_model import Perceptron
perceptron = Perceptron()
perceptron.fit(xtrain.reshape(-1, 1), ytrain.reshape(-1, 1))
accuracy = round(perceptron.score(xtrain.reshape(-1, 1), ytrain.reshape(-1, 1)) * 100, 2)
print (accuracy)
```

```python
y_pred = perceptron.predict(xtest.reshape(-1, 1))
print (ytest)
print (y_pred)
```
